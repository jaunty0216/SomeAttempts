```python
from IPython import get_ipython
from IPython.display import display
# %%
!sudo apt-get update
!sudo apt-get install fonts-noto-cjk
# %%
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = "/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"
font_prop = fm.FontProperties(fname=font_path, size=10)

def draw_new_process():
    # 定義加工活動的詳細內容 (New tasks)
    task_labels = {
        "A_1": "A_1\n構思草圖",
        "A_2": "A_2\n擬定設計圖",
        "A_3": "A_3\n將設計圖轉到紙模上",
        "A_4": "A_4\n將紙模的稿繪製到鋼板上",
        "B_1": "B_1\n將鋼板裁剪下來",
        "B_2": "B_2\n將鋼板折成船的形狀",
        "B_3": "B_3\n裝飾作為棚子的鋼板",
        "B_4": "B_4\n將鋼板折成棚子的形狀",
        "C_1": "C_1\n將兩塊鋼板合併在一起",
        "C_2": "C_2\n修正船頭",
        "C_3": "C_3\n安裝馬達電路系統",
        "C_4": "C_4\n完成",
    }

    # 定義活動的依賴關係 (New dependencies)
    tasks = {
        "A_1": [],
        "A_2": ["A_1"],
        "A_3": ["A_2"],
        "A_4": ["A_3"],
        "B_1": ["A_4"], # Assuming B_1 depends on A_4
        "B_2": ["B_1"],
        "B_3": ["A_4"], # Assuming B_3 depends on A_4
        "B_4": ["B_3"],
        "C_1": ["B_2", "B_4"], # Assuming C_1 depends on both B_2 and B_4
        "C_2": ["C_1"],
        "C_3": ["C_2"],
        "C_4": ["C_3"],
    }

    # 建立有向圖
    G = nx.DiGraph()
    for task, dependencies in tasks.items():
        G.add_node(task, label=task_labels[task])
        for dep in dependencies:
            G.add_edge(dep, task)

    # 繪製圖表
    width, height = 20/2.54, 22/2.54
    plt.figure(figsize=(width,height))  # 調整圖表大小
    pos = nx.spring_layout(G, k=0.3, seed=789)  # 調整節點間距
    labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=1550,
            node_color="lightblue", font_size=0.5, font_weight="bold", edge_color="gray",
            node_shape='s')

    # 設定中文字體
    for text in plt.gca().texts:
        text.set_fontproperties(font_prop)

    plt.title("新加工流程圖", fontproperties=font_prop)

    plt.margins(0.1)  # 調整圖表邊界

    plt.show()

# 執行繪製流程圖的函數
draw_new_process()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

font_path = "/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"
font_prop = fm.FontProperties(fname=font_path, size=10)

def draw_new_process_with_seed(seed_value):
    # 定義加工活動的詳細內容
    task_labels = {
        "A_1": "A_1\n構思草圖",
        "A_2": "A_2\n擬定設計圖",
        "A_3": "A_3\n將設計圖轉到紙模上",
        "A_4": "A_4\n將紙模的稿繪製到鋼板上",
        "B_1": "B_1\n將鋼板裁剪下來",
        "B_2": "B_2\n將鋼板折成船的形狀",
        "B_3": "B_3\n裝飾作為棚子的鋼板",
        "B_4": "B_4\n將鋼板折成棚子的形狀",
        "C_1": "C_1\n將兩塊鋼板合併在一起",
        "C_2": "C_2\n修正船頭",
        "C_3": "C_3\n安裝馬達電路系統",
        "C_4": "C_4\n完成",
    }

    # 定義活動的依賴關係
    tasks = {
        "A_1": [],
        "A_2": ["A_1"],
        "A_3": ["A_2"],
        "A_4": ["A_3"],
        "B_1": ["A_4"],
        "B_2": ["B_1"],
        "B_3": ["A_4"],
        "B_4": ["B_3"],
        "C_1": ["B_2", "B_4"],
        "C_2": ["C_1"],
        "C_3": ["C_2"],
        "C_4": ["C_3"],
    }

    # 建立有向圖
    G = nx.DiGraph()
    for task, dependencies in tasks.items():
        G.add_node(task, label=task_labels[task])
        for dep in dependencies:
            G.add_edge(dep, task)

    # 繪製圖表
    width, height = 20/2.54, 22/2.54
    plt.figure(figsize=(width,height))
    pos = nx.spring_layout(G, k=0.3, seed=seed_value)
    labels = nx.get_node_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, labels=labels, node_size=1550,
            node_color="lightblue", font_size=0.5, font_weight="bold", edge_color="gray",
            node_shape='s')

    # 設定中文字體
    for text in plt.gca().texts:
        text.set_fontproperties(font_prop)

    plt.title(f"新加工流程圖 (Seed: {seed_value})", fontproperties=font_prop)
    plt.margins(0.1)
    plt.show()

# 嘗試不同的 seed 值
seed_values_to_try = [42, 123, 789, 1000, 5555]

for seed in seed_values_to_try:
    draw_new_process_with_seed(seed)
```
